```python
corr = df.corr()[abs(df.corr()) >= .01].stack().reset_index()

corr2 = corr[corr[0] != 1].sort_values(by = 0).drop_duplicates(subset = 0)

corr3 = corr2[abs(corr2[0]) >= .5]

correlation = []

for value in corr3[0]:
    if value > 0:
        correlation.append('pos_corr')
    else:
        correlation.append('neg_corr')
        
corr3['corr'] = correlation

corr3
```
